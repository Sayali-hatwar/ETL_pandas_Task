## 1. Essential Properties (Attributes)
  Properties tell you something about the data without changing it. You don't use parentheses () with these.

  | Property | Description | SQL/Power BI Equivalent |
| :--- | :--- | :--- |
| **`.shape`** | Returns a tuple (rows, columns). | Table Row/Col count |
| **`.columns`** | Returns the list of column names. | Header Row |
| **`.index`** | Returns the row labels (usually 0, 1, 2...). | Primary Key/Index |
| **`.dtypes`** | Shows the data type of each column. | Schema / Data Category |
| **`.size`** | Total number of elements (rows × columns). | Total Cell Count |
| **`.values`** | Converts the data into a NumPy array. | Raw Data Array |

## 2. Essential Methods
Methods perform an action or calculation on the data. These always use parentheses ().

**A. Data Exploration**
* **.head(n) / .tail(n):** View the first or last (n) rows.
* **.info():** A "health check" for your data. Shows non-null counts and memory usage.
* **.describe():** Generates summary statistics (mean, std, min, max, quartiles).

-Power BI Equivalent: Column Statistics in Power Query.

```python
pop_2010_2019_df.head(10)
```

```python
pop_2010_2019_df.tail(10)
```

```python
pop_2010_2019_df.info()
```

```python
pop_2010_2019_df.describe()
```

**B. Data Selection & Filtering**
* **.loc[]:** Access rows and columns by label (e.g., df.loc[0, 'Name']). --- only works on index col..far left col

* **.iloc[]:** Access rows and columns by integer position (e.g., df.iloc[0, 1]).

* **.drop():** Removes specified rows or columns.

```python
pop_2010_2019_df.loc[226]
```

```python
print(pop_2010_2019_df.iloc[1])
print("=="*12)
print(pop_2010_2019_df.iloc[0])
```

**C. Data Cleaning**
* **.isnull() / .notnull():** Returns a boolean mask for missing values.

* **.fillna(value):** Replaces NaN values with a specific value.

--SQL Equivalent: COALESCE(column, value).

* **.dropna():** Deletes rows/columns containing missing values.

* **.rename():** Changes column or index names.

```python
pop_2010_2019_df.isnull()
```

```python
pop_2010_2019_df.isnull().any()
```

```python
pop_2010_2019_df.isnull().any(axis=1)
```

**D. Data Transformation**
* **.groupby():** Groups data by a column to perform aggregations.

--SQL Equivalent: GROUP BY.

* **.sort_values():** Sorts the data by one or more columns.

--SQL Equivalent: ORDER BY.

* **.apply():** Applies a function (like a Lambda) to every element in a column.
